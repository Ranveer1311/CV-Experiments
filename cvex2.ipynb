```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow as cv2
```

```python
img=cv.imread('/content/j.png')
cv2(img)
```

```python
#erosion
kernel = np.ones((5,5),np.uint8)
erode=cv.erode(img,kernel ,iterations=1)
cv2(erode)
```

```python
#dialation
dilate=cv.dilate(img,kernel,iterations=1)
cv2(dilate)
```

```python
#opening and closing
img1=cv.imread('/content/opening.png')
img2=cv.imread('/content/closing.png')
cv2(img1)
cv2(img2)
```

```python
opening=cv.dilate(cv.erode(img1,kernel ,iterations=1),kernel,iterations=1)
cv2(opening)
```

```python
opening=cv.erode(cv.dilate(img2,kernel ,iterations=1),kernel,iterations=1)
cv2(opening)
```

```python
img3=cv.imread('/content/circle.png')
img4=cv.imread('/content/finger2.PNG')
cv2(img3)
cv2(img4)
```

```python
#erosion
kernel = np.ones((3,3),np.uint8)
erode1=cv.erode(img3,kernel ,iterations=5)
erode2=cv.erode(img4,kernel ,iterations=3)
cv2(erode1)
cv2(erode2)
```

```python
#dilate
dilate1=cv.dilate(img3,kernel ,iterations=1)
dilate2=cv.dilate(img4,kernel ,iterations=1)
cv2(dilate1)
cv2(dilate2)
```
